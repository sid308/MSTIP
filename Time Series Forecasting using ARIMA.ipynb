```python
#ignore the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.arima.model import ARIMA
```

```python
#load the data and split it into a training data set and a test set
url = "https://raw.githubusercontent.com/facebook/prophet/main/examples/example_air_passengers.csv"
data = pd.read_csv(url, header=0, parse_dates=[0], index_col=0,)
train = data.iloc[:-12] 
test = data.iloc[-12:]
```

```python
#fit an ARIMA model
arima_model = ARIMA(train, order=(5,1,0))
arima_result = arima_model.fit()

#i've cleared the output of this cell because they were warnings showing off my per
```

```python
#forecasts the next 12 months of airline passengers
arima_forecast = arima_result.forecast(steps=12)
```

```python
#plot the forecast
plt.figure(figsize=(10, 5))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(arima_forecast, label='ARIMA Forecast')
plt.legend()
plt.title('ARIMA Forecasting')
plt.show()
```
